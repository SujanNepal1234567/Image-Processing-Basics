##HOG feature Vector 


```python
from skimage.io import imread, imshow
from skimage.feature import hog
from skimage import exposure, data


```

```python
import matplotlib.pyplot as plt 
from skimage import color
```

```python
rocket_image = data.rocket()
grayscale =color.rgb2gray(rocket_image)
plt.imshow(rocket_image, cmap="gray")
plt.title("Image")
plt.axis("off")
```

```python
MC= True
# MC = False

hogfv, hog_image = hog(rocket_image,orientations=9,pixels_per_cell=(16,16),cells_per_block=(2,2), visualize=True, channel_axis=-1)
```

```python
hog_image_rescaled = exposure.rescale_intensity(hog_image, in_range=(0,5))
imshow(hog_image_rescaled)
```
